```python
import pandas as pd 
from rdkit import Chem
import numpy as np
```

```python
tge = pd.read_csv('../../../data/targetome_extended-01-23-25.csv')
tge_meta = pd.read_csv('../../../data/targetome_extended_drugs-01-23-25.csv')
# /home/teddy/local/TKG-DTI/workflow/scripts/retrieve_smiles.ipynb
```

```python
def get_smiles_from_inchi(inchis, verbose=True):

    inchis = np.unique(inchis).tolist() 

    inchi2smiles = {} 
    for i, inchi in enumerate(inchis):    
        if verbose:
            print(f"progress: {i}/{len(inchis)}", end="\r")
        mol = Chem.MolFromInchi(inchi)
        if mol is not None:
            smiles_string = Chem.MolToSmiles(mol)
            inchi2smiles[inchi] = smiles_string
        else:
            inchi2smiles[inchi] = None

    return inchi2smiles

smiles2inchi = get_smiles_from_inchi(tge_meta.inchi.values)
```
